# Creating an EntitySet

```python
# initialize an EntitySet
es = ft.EntitySet(id="customer_data")
```

```python
es.plot()
```

```python
# add elements
'''
The index parameter specifies the column that uniquely identifies rows in the dataframe
The time_index parameter tells Featuretools when the data was created.
The variable_types parameter indicates that “product_id” should be interpreted as a Categorical variable, even though it just an integer in the underlying data.
'''
es = es.entity_from_dataframe(entity_id="transactions",
                              dataframe=transactions_df,
                              index="transaction_id", # unique identifier
                              time_index="transaction_time", # Name of the variable containing time data
                              variable_types={"product_id": ft.variable_types.Categorical,
                                              "zip_code": ft.variable_types.ZIPCode})
```

```python
es.plot()
```

```python
es["transactions"].variables
```

```python
es = es.entity_from_dataframe(entity_id="products",
                              dataframe=products_df,
                              index="product_id")
es.plot()
```

# Adding a Relationship

```python
# Each product has multiple transactions associated with it, so it is called it the parent entity, 
# while the transactions entity is known as the child entity.
new_relationship = ft.Relationship(parent_variable=es["products"]["product_id"],
                                   child_variable=es["transactions"]["product_id"])
```

```python
es = es.add_relationship(new_relationship)
es.plot()
```

# Creating entity from existing table

```python
# Create a new entity and relationship from unique values of an existing variable.
es = es.normalize_entity(base_entity_id="transactions", # build new entity based on "transaction" entity
                         new_entity_id="sessions", # name of the new entity
                         index="session_id", 
                         make_time_index="session_start",
                         additional_variables=["device", "customer_id", "zip_code", "session_start", "join_date"])# It removed “device”, “customer_id”, “zip_code” and “join_date” from “transactions” and created a new variables in the sessions entity
```

```python
es.plot()
```

```python
es = es.normalize_entity(base_entity_id="sessions",
                         new_entity_id="customers",
                         index="customer_id",
                         make_time_index="join_date",
                         additional_variables=["zip_code", "join_date"])
```

```python
es.plot()
'''
each customer_id(parent entity) has multiple sessions(child), each sesson(parent) has multiple transactions(child entity)
parent entity is more like a summary of info/reference of the child entity
'''
```

```python
feature_matrix, feature_defs = ft.dfs(entityset=es, target_entity="sessions")
```

# Deep Feature Synthesis

```python
es = ft.demo.load_mock_customer(return_entityset=True)
```

```python
es.plot()
```

```python
# depth = 1 
feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_entity="customers",
                                      agg_primitives=["count"], # count the No. of sessions associated with the customer_id
                                      trans_primitives=["month"], # transform datetime to the Month 
                                      max_depth=1)
feature_matrix
```

```python
# depth = 2
feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_entity="customers",
                                      agg_primitives=["count"], # count the No. of sessions associated with the customer_id
                                      trans_primitives=["month"], # transform datetime to the Month 
                                      max_depth=2) # The No. of children entities
feature_matrix
```
